```python
import pandas as pd
import matplotlib.pylab as plt

df_cov = pd.read_csv('covid19_confirmed_global.csv')
df_cov
```

```python
# 美國
df_cov_us = df_cov[df_cov['Country/Region'] == 'US']

# 新增
df_cov_us.loc['2','1/22/20':] = 0 
df_cov_us.loc['2','Country/Region']='us_add'
#df_cov_ger

#df_cov_ger.iloc[0,5:103]
for i in range(6,102):
    df_cov_us.iloc[1,i] = df_cov_us.iloc[0,i] - df_cov_us.iloc[0,i-1]
# 整數
df_cov_us.loc[:,'1/22/20':] = df_cov_us.loc[:,'1/22/20':].astype(int)
# 另存
df_cov_us.to_csv('./covid19_confirmed_Us.csv')
df_cov_us
```

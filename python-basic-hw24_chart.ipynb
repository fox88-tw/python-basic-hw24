```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 正常顯示 中文 , 將預設字體 指定為 微軟雅黑體
plt.rcParams['font.family'] = ['Microsoft YaHei'] 
# 正常顯示 負號
plt.rcParams['axes.unicode_minus']=False 
plt.figure(figsize=(10,16))

df = pd.read_csv('covid19_index.csv', index_col=0)
df.index = pd.to_datetime(df.index)

df
```

```python
#df['T_index'].dropna()
#mask = np.isfinite(df)
#mask['T_index']
#df['mask']
df['usd_index'].dropna().plot()
df['usd_index'].dropna()
```

```python
fig, ax1 = plt.subplots()
#ax2 = ax1.twinx()
ax1.bar(df.index, df['Global_add'], color='r')
df['T_index'].dropna().plot(kind='line', color= '#304547', secondary_y=True , grid=False)



ax1.set_title('全球 Covid-19 新增病例 vs 貨幣指數',size= 19)
ax1.set_ylabel('確診人數',color= 'r',size= 14)
ax1.set_xlabel('日期',size= 14)
ax1.tick_params(axis='y',colors='r')

ax1.right_ax.set_ylabel('貨幣指數',size= 14)

fig.autofmt_xdate(rotation=45)

plt.savefig('Global.png',dpi=150,bbox_inches='tight')
```

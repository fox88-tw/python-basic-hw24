```python
import pandas as pd
import matplotlib.pylab as plt

df_cov = pd.read_csv('covid19_confirmed_global.csv')
df_usd = pd.read_csv('us_index-01.csv')
df_eur = pd.read_csv('eur_index-01.csv')
df_jpy = pd.read_csv('jpy_index-01.csv')
```

```python
#df_cov = DataFrame.drop
df_cov2 = df_cov.drop(columns=['Unnamed: 0', 'Province/State', 'Lat', 'Long'])
df_cov3 = df_cov2.drop(index=range(0,264))

df_cov3.rename(columns={'Country/Region':''}, inplace=True)
df_cov3 = df_cov3.set_index('').T
df_cov3.index.name = 'date'
df_cov3
#df_cov3.plot(kind='line')
#plt.show()
```

```python
df_usd1 = df_usd.set_index('date')
df_usd1.rename(columns={'price':'usd_index'}, inplace=True)

df_glo = df_cov3.merge(df_usd1, how = 'left', on = 'date')

df_eur1 = df_eur.set_index('date')
df_eur1.rename(columns={'price':'eur_index'}, inplace=True)

df_glo = df_glo.merge(df_eur1, how = 'left', on = 'date')


df_jpy1 = df_jpy.set_index('date')
df_jpy1.rename(columns={'price':'jpy_index'}, inplace=True)

df_glo = df_glo.merge(df_jpy1, how = 'left', on = 'date')

df_glo
```

```python
df_glo['T_index'] = df_glo[['usd_index', 'eur_index', 'jpy_index']].sum(1,False)

df_glo.to_csv('./covid19_index.csv')

df_glo
```

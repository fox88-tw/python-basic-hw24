```python
import pandas as pd
import matplotlib.pylab as plt

df_cov = pd.read_csv('covid19_confirmed_global.csv', index_col=0)
df_usd = pd.read_csv('us_index-01.csv', index_col=0)
df_eur = pd.read_csv('eur_index-01.csv', index_col=0)
df_jpy = pd.read_csv('jpy_index-01.csv', index_col=0)
```

```python
# 刪除行列
df_cov2 = df_cov.drop(columns=['Province/State', 'Lat', 'Long'])
df_cov3 = df_cov2.drop(index=range(0,264))

df_cov3.rename(columns={'Country/Region':''}, inplace=True)
df_cov3 = df_cov3.set_index('').T
df_cov3.index.name = 'date'
df_cov3
#df_cov3.plot(kind='line')
#plt.show()
```

```python
import pandas as pd
import matplotlib.pylab as plt

df_cov_us = pd.read_csv('covid19_confirmed_Us.csv', index_col=0)
df_cov_jp = pd.read_csv('covid19_confirmed_Jp.csv', index_col=0)
df_cov_ge = pd.read_csv('covid19_confirmed_Germany.csv', index_col=0)
```

```python
# US
# 刪除行列
df_cov2_us = df_cov_us.drop(columns=['Unnamed: 0.1', 'Province/State', 'Lat', 'Long'])
df_cov2_us.rename(columns={'Country/Region':''}, inplace=True)
df_cov2_us = df_cov2_us.set_index('').T
df_cov2_us.index.name = 'date'
df_cov2_us
#df_cov3.plot(kind='line')
#plt.show()
```

```python
# JP
# 刪除行列
df_cov2_jp = df_cov_jp.drop(columns=['Unnamed: 0.1', 'Province/State', 'Lat', 'Long'])
df_cov2_jp.rename(columns={'Country/Region':''}, inplace=True)
df_cov2_jp = df_cov2_jp.set_index('').T
df_cov2_jp.index.name = 'date'
df_cov2_jp
#df_cov3.plot(kind='line')
#plt.show()
```

```python
# Germany
# 刪除行列
df_cov2_ge = df_cov_ge.drop(columns=['Unnamed: 0.1', 'Province/State', 'Lat', 'Long'])
df_cov2_ge.rename(columns={'Country/Region':''}, inplace=True)
df_cov2_ge = df_cov2_ge.set_index('').T
df_cov2_ge.index.name = 'date'
df_cov2_ge
#df_cov3.plot(kind='line')
#plt.show()
```

```python
# 依照 index 合併 數據 : covid 全球+各國
df_glo = df_cov3.merge(df_cov2_us, how = 'left', on = 'date')
df_glo = df_glo.merge(df_cov2_ge, how = 'left', on = 'date')
df_glo = df_glo.merge(df_cov2_jp, how = 'left', on = 'date')
df_glo
```

```python
# 依照 index 合併 數據 : covid + index
df_usd.rename(columns={'price':'usd_index'}, inplace=True)
df_glo = df_glo.merge(df_usd, how = 'left', on = 'date')

df_eur.rename(columns={'price':'eur_index'}, inplace=True)
df_glo = df_glo.merge(df_eur, how = 'left', on = 'date')

df_jpy.rename(columns={'price':'jpy_index'}, inplace=True)
df_glo = df_glo.merge(df_jpy, how = 'left', on = 'date')

df_glo
```

```python
df_glo['T_index'] = df_glo[['usd_index', 'eur_index', 'jpy_index']].sum(1,False)

df_glo.to_csv('./covid19_index.csv')

df_glo
```
